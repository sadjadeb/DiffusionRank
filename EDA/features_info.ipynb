```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
project_root = '/mnt/data/sajadeb/Gen_LTR'

X_real_cal = np.load(os.path.join(project_root, 'data', 'california', 'X_num_train.npy'))
X_real_letor = np.load(os.path.join(project_root, 'data', 'MQ2007', 'npy', 'Fold1', 'X_num_train.npy'))
X_real_mslr = np.load(os.path.join(project_root, 'data', 'MSLR-WEB30K', 'npy', 'Fold1', 'X_num_train.npy'))

n_features_cal = X_real_cal.shape[1]
n_features_letor = X_real_letor.shape[1]
n_features_mslr = X_real_mslr.shape[1]
```

```python
def calculate_statistics(X_real, n_features):
    stats_dict = {
        'Mean': np.mean(X_real, axis=0),
        'Std': np.std(X_real, axis=0),
        'Min': np.min(X_real, axis=0),
        'Max': np.max(X_real, axis=0),
        'Skewness': stats.skew(X_real, axis=0),
        'Kurtosis': stats.kurtosis(X_real, axis=0),
        'Unique Values': [len(np.unique(X_real[:, i])) for i in range(n_features)]
    }
    stats_df = pd.DataFrame(stats_dict, index=[f'Feature {i}' for i in range(n_features)])
    return stats_df
```

```python
calculate_statistics(X_real_cal, n_features_cal)
```

```python
calculate_statistics(X_real_letor, n_features_letor)
```

```python
calculate_statistics(X_real_mslr, n_features_mslr)
```
