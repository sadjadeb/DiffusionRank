## Distribution of labels in training sets per dataset

```python
y_train_mq2007 = np.load(os.path.join(project_root, 'data', 'MQ2007', 'by_fraction', 'Fold1', 'k1.0', 'y_train.npy'))
y_train_mq2008 = np.load(os.path.join(project_root, 'data', 'MQ2008', 'by_fraction', 'Fold1', 'k1.0', 'y_train.npy'))
y_train_mslr10 = np.load(os.path.join(project_root, 'data', 'MSLR-WEB10K', 'by_fraction', 'Fold1', 'k1.0', 'y_train.npy'))
y_train_mslr30 = np.load(os.path.join(project_root, 'data', 'MSLR-WEB30K', 'by_fraction', 'Fold1', 'k1.0', 'y_train.npy'))
```

```python
unique_labels_mq2007, labels_counts_mq2007 = np.unique(y_train_mq2007, return_counts=True)
unique_labels_mq2008, labels_counts_mq2008 = np.unique(y_train_mq2008, return_counts=True)
unique_labels_mslr10, labels_counts_mslr10 = np.unique(y_train_mslr10, return_counts=True)
unique_labels_mslr30, labels_counts_mslr30 = np.unique(y_train_mslr30, return_counts=True)
```

```python
plt.figure(figsize=(12, 8), dpi=150)
plt.subplot(2, 2, 1)
plt.bar(unique_labels_mq2007, labels_counts_mq2007, width=0.4, color="#4C72B0")
plt.title('MQ2007 Training Set Labels')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.xticks(unique_labels_mq2007)
plt.subplot(2, 2, 2)
plt.bar(unique_labels_mq2008, labels_counts_mq2008, width=0.4, color="#DD8452")
plt.title('MQ2008 Training Set Labels')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.xticks(unique_labels_mq2008)
plt.subplot(2, 2, 3)
plt.bar(unique_labels_mslr10, labels_counts_mslr10, color="#55A868")
plt.title('MSLR-WEB10K Training Set Labels')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.xticks(unique_labels_mslr10)
plt.subplot(2, 2, 4)
plt.bar(unique_labels_mslr30, labels_counts_mslr30, color="#C44E52")
plt.title('MSLR-WEB30K Training Set Labels')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.xticks(unique_labels_mslr30)
plt.tight_layout()
# plt.show()
plt.savefig(os.path.join(project_root, 'EDA', 'output', 'datasets_labels_distribution.png'))
```
